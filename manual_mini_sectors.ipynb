```python
import fastf1
from fastf1 import plotting
from fastf1.plotting import setup_mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the qualifying session for the 2022 Spanish GP
session = fastf1.get_session(2022, 'Spanish Grand Prix', 'Q')
session.load()

# Get Verstappen fastest lap
ver_lap = session.laps.pick_driver('VER').pick_fastest()
tel = ver_lap.get_telemetry()
```

```python

# Extract telemetry data
x = tel['X']
y = tel['Y']       
dist = tel['Distance']

# Manually define sector boundaries (in meters)
sector_bounds = [0,750,1250,1600,1870,2000,2200,2450,2600,2800,3000,3370,3900,4400,4636.84]
print(dist.max())

# Assign telemetry points to sectors
sector_idx = np.digitize(dist, sector_bounds) - 1

# Plot
plotting.setup_mpl()

plt.figure(figsize=(10, 7))
for i in range(len(sector_bounds) - 1):
    mask = sector_idx == i
    plt.plot(x[mask], y[mask], label=f'Sector {i+1}')

plt.title('Max Verstappen Fastest Lap (Q) - Spain 2022\nTrack divided into 10 manual sectors')
plt.axis('equal')
plt.legend()
plt.show()
```

# British Grand Prix (Silverstone)

```python
session = fastf1.get_session(2022, 'British Grand Prix', 'Q')
session.load()

# Get Verstappen’s fastest lap
ver_lap = session.laps.pick_driver('VER').pick_fastest()
tel = ver_lap.get_telemetry()
```

```python
# Extract telemetry data
x = tel['X']       # x-coordinate of car position
y = tel['Y']       # y-coordinate of car position
dist = tel['Distance']

print(dist.max())

# Manually define sector boundaries (in meters)
sector_bounds = [0,350,800,1150,1830,2300,2930,3170,3600,4200,4900,5130,5400,5650,dist.max()]
# Assign telemetry points to sectors
sector_idx = np.digitize(dist, sector_bounds) - 1

#label the sector
# 1 - slow, 2 - medium, 3 - fast, 4 - straight
sector_labels = (4,3,1,4,2,4,3,4,3,4,2,4,1,4)

# --- Color-blind–safe palette (Okabe-Ito) ---
colors = {
    1: "#E69F00",  # orange
    2: "#56B4E9",  # sky blue
    3: "#009E73",  # bluish green
    4: "#F0E442",  # yellow
}

# --- Label mapping ---
label_names = {
    1: "slow",
    2: "medium",
    3: "fast",
    4: "straight",
}

# --- Plot ---
plotting.setup_mpl()
plt.figure(figsize=(10, 7))

for i in range(len(sector_bounds) - 1):
    mask = sector_idx == i
    label = sector_labels[i]
    plt.plot(x[mask], y[mask], color=colors[label], linewidth=2.5, label=label_names[label])

plt.title("Max Verstappen Fastest Lap (Q) - Spain 2022\n10 Manual Sectors with Random Speed Labels")
plt.axis('equal')

# Deduplicate legend entries (so each label appears once)
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), title="Speed Category")

plt.show()
```
